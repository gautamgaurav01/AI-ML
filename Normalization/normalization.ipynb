# Scaling and Normalization


```python
# Import required libraries
import pandas as pd
import numpy as np
from scipy import stats
from mlxtend.preprocessing import minmax_scaling
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
kickstarters_2017 = pd.read_csv("./ks-projects-201612.csv", encoding="latin1", low_memory=False)

# Strip whitespace from column names to avoid AttributeError
kickstarters_2017.columns = kickstarters_2017.columns.str.strip()

# Set random seed
np.random.seed(0)

print("Setup Complete")
```

## 1) Practice Scaling

```python
# Select the goal column, coerce non-numeric values (e.g. date strings) to NaN, drop them
original_goal_data = pd.DataFrame(
    pd.to_numeric(kickstarters_2017['goal'], errors='coerce')
).dropna()

# Scale the goals from 0 to 1
scaled_goal_data = minmax_scaling(original_goal_data, columns=['goal'])

print('Original data\nPreview:\n', original_goal_data.head())
print('Minimum value:', original_goal_data['goal'].min())
print('Maximum value:', original_goal_data['goal'].max())
print('_' * 30)

print('\nScaled data\nPreview:\n', scaled_goal_data.head())
print('Minimum value:', scaled_goal_data['goal'].min())
print('Maximum value:', scaled_goal_data['goal'].max())
```

## 2) Practice Normalization

```python
# Convert pledged to numeric, coercing any non-numeric values to NaN
kickstarters_2017['pledged'] = pd.to_numeric(kickstarters_2017['pledged'], errors='coerce')

# Get only positive pledged values (also drops NaN via the > 0 comparison)
positive_pledges = kickstarters_2017.loc[kickstarters_2017['pledged'] > 0, 'pledged'].dropna()

# Normalize the pledged data using Box-Cox transformation
normalized_pledges = pd.Series(
    stats.boxcox(positive_pledges)[0],
    name='pledged',
    index=positive_pledges.index
)

print('Original data\nPreview:\n', positive_pledges.head())
print('Minimum value:', positive_pledges.min())
print('Maximum value:', positive_pledges.max())
print('_' * 30)

print('\nNormalized data\nPreview:\n', normalized_pledges.head())
print('Minimum value:', normalized_pledges.min())
print('Maximum value:', normalized_pledges.max())
```

```python
# Plot normalized data
ax = sns.histplot(normalized_pledges, kde=True)
ax.set_title("Normalized pledged data")
plt.show()
```

## Conclusion

The normalized `usd_pledged_real` and `pledged` columns look mostly similar after normalization because both represent pledged money amounts.

However, there may be slight differences because:

- `usd_pledged_real` contains values converted to USD
- `pledged` contains values in different original currencies
